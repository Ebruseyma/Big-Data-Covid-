```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Purpose of this notebook is to get updated data from the internet and then filter data and save to modified_data folder with name 'corona_new'

# Get data to Internet
covid = pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv')
del covid['iso_code']
del covid['date']
del covid['positive_rate']


# Groupby data 'location' and aggregate important columns 

covid_new = covid.groupby(['location']).agg({'total_cases':'last',\
                                             'total_deaths': 'last',\
                                             'total_cases_per_million': 'last',\
                                             'total_deaths_per_million': 'last', \
                                             'reproduction_rate': 'mean', \
                                             'icu_patients': 'last',\
                                             'icu_patients_per_million': 'last' ,\
                                             'hosp_patients': 'last', \
                                             'hosp_patients_per_million': 'last',\
                                             'total_vaccinations': 'last', \
                                             'total_vaccinations_per_hundred': \
                                             'last', 'stringency_index': 'mean',\
                                             'median_age': 'mean', \
                                             'aged_65_older': 'last', \
                                             'aged_70_older': 'last', \
                                             'extreme_poverty': 'last',\
                                             'cardiovasc_death_rate': 'last', \
                                             'diabetes_prevalence': 'last', \
                                             'female_smokers': 'mean',\
                                             'male_smokers': 'mean', \
                                             'handwashing_facilities': 'last', \
                                             'hospital_beds_per_thousand':'mean',\
                                             'life_expectancy': 'last', \
                                             'human_development_index': 'mean',\
                                             })

# Print unfiltered and grouped info
print(covid.info())


# Delete some features due to not enough information may be in future it would be added if enough data is added from internet
del covid_new['icu_patients']
del covid_new['icu_patients_per_million']
del covid_new['hosp_patients']
del covid_new['hosp_patients_per_million']
del covid_new['total_vaccinations']
del covid_new['total_vaccinations_per_hundred']
del covid_new['handwashing_facilities']

# Print filtered info
print('Filtered Data')
print(covid_new.info())

covid_new.to_csv('modified_data/table_corona_new.csv')
```
